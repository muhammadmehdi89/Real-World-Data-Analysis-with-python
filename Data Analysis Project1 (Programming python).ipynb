```python
female_parti = ath_rg[(ath_rg.Sex=='F') & (ath_rg.Season=='Summer')][['Sex','Year']]
female_parti = female_parti.groupby('Year').count().reset_index()
female_parti.head()
```

```python
womenOlymp = ath_rg[(ath_rg.Sex=='F') & (ath_rg.Season=='Summer')]
```

```python
sns.set(style='darkgrid')
plt.figure(figsize=(20,10))
sns.countplot(x='Year',data=womenOlymp,palette='Spectral')
plt.title('Women Participation')
```

```python
part = womenOlymp.groupby('Year')['Sex'].value_counts()
plt.figure(figsize=(20,10))
part.loc[:,'F'].plot()
plt.title('Plot of Female Athletes over time')
```
